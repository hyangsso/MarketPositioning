```python
# 상품 정보 url 수집

from selenium import webdriver
from bs4 import BeautifulSoup
import requests
from tqdm.notebook import tqdm
import pandas as pd


url_list = []

keyword = ['seaweed', 'chip']

for i in range(1,6):
    url = "https://www.amazon.com/s?k=" + keyword[0] + '+' + keyword[1] + '&rh=n%3A16322721&ref=nb_sb_noss&' + 'page=' + str(i)
    path = "C:/Users/user/chromedriver_win32/chromedriver"
    driver = webdriver.Chrome(path)
    driver.get(url)
    page = driver.page_source
    soup = BeautifulSoup(page, 'html.parser')
    
    list_ = soup.select("div.a-section.a-spacing-none.a-spacing-top-small > h2 > a.a-link-normal.a-text-normal")
    
    for l in list_:
        url_ = l.attrs['href']
        url_list.append("https://www.amazon.com" + url_)
        
    driver.close()
```

```python
url_list = pd.DataFrame( data = {'url_list': url_list})
```

```python
url_list = url_list[~url_list['url_list'].str.contains("sp_mtf")]
```

```python
# 상품 정보 수집

'''
url_list = ["https://www.amazon.com/C-weed-Korean-Crunchy-Roasted-Seaweed/dp/B07TD6YWH3/ref=sr_1_1?dchild=1&keywords=bugak&qid=1597199084&sr=8-1",
            "https://www.amazon.com/Annie-Chuns-Baked-Seaweed-Crisps/dp/B00A1FINLM/ref=sr_1_2?dchild=1&keywords=bugak&qid=1597203294&sr=8-2", 
            "https://www.amazon.com/Korea-Healthy-Jeonju-Blossom-Seaweed/dp/B07L42HR8Z/ref=sr_1_3?dchild=1&keywords=bugak&qid=1597203226&sr=8-3",
            "https://www.amazon.com/C-Weed-Korean-Seaweed-Almonds-Resealable/dp/B07GK6GF11/ref=sr_1_9?dchild=1&keywords=bugak&qid=1597203238&sr=8-9",
            "https://www.amazon.com/SeaSnax-Chomperz-Crunchy-Seaweed-BARBECUE/dp/B00W4TDUJ8/ref=sr_1_37?dchild=1&keywords=seaweed%2Bsnack%2Blaver&qid=1597280815&sr=8-37&th=1", 
            "https://www.amazon.com/EYN-Laver-chili-snack/dp/B01LCNV0T8/ref=sr_1_60?dchild=1&keywords=seaweed+snack+laver&qid=1597281161&sr=8-60",
            "https://www.amazon.com/Seaweed-Crisps-Flavor-Crunch-Non-GMO/dp/B07GB893LM/ref=sr_1_64?dchild=1&keywords=seaweed+snack+laver&qid=1597281161&sr=8-64",
            "https://www.amazon.com/Dongwon-Crispy-Laver-Snack-Roasted/dp/B07Q1XH4R7/ref=sr_1_71?dchild=1&keywords=seaweed+snack+laver&qid=1597281161&sr=8-71", 
            "https://www.amazon.com/Dongwon-Crispy-Laver-Snack-Shrimp/dp/B07PZS6TTT/ref=sr_1_72?dchild=1&keywords=seaweed+snack+laver&qid=1597281161&sr=8-72"]
'''
title = []
categories = []
flavors = []
size = []
features = []
price = []

for url in tqdm(url_list['url_list']):
    path = "C:/Users/user/chromedriver_win32/chromedriver"
    driver = webdriver.Chrome(path)
    driver.get(url)
    page = driver.page_source
    soup = BeautifulSoup(page, 'html.parser')
    
    try:
        title.append(soup.select("#productTitle")[0].text.strip('\n'))
    except:
        title.append('x')
    
    try:
        soup.se
        
    
    try:
        size.append(soup.select('span.selection')[0].text.strip('\n'))
    except:
        size.append('x')
    
    try:
        for li in soup.select("#wayfinding-breadcrumbs_container ul.a-unordered-list")[0].findAll("li"):
            category = li.text.strip('\n')
        categories.append(category)
    except:
        categories.append('x')
    
    try:
        for li in soup.select("#feature-bullets ul.a-unordered-list")[0].findAll('li'):
            feature = li.text.strip('\n')
        features.append(feature)
    except:
        features.append('x')
    
    try:
        price.append(soup.select("span.a-size-small.a-color-price")[0].text.strip('\n'))
    except:
        price.append('x')
     
    driver.close()
```

```python
amazon_product_result = pd.DataFrame( data = {'title': title, 'categories': categories, 'size': size, 'features' : features, 'price' : price}) ; amazon_product_result  
```

```python
# 상품 리뷰 수집

import re
import time


star_ratings = []
comment_authors = []
comment_titles = []
comments = []
comment_dates = []

for url in tqdm(url_list['url_list']):
    path = "C:/Users/user/chromedriver_win32/chromedriver"
    driver = webdriver.Chrome(path)
    driver.get(url)
    
    try:
        content = driver.find_element_by_css_selector("span.a-size-small a.a-link-normal")
        content.click()
        
        page_source = driver.page_source
        current_url = driver.current_url
        soup = BeautifulSoup(page_source,"html.parser")
        time.sleep(0.5)

        count = soup.find("span", {"data-hook": "cr-filter-info-review-count"}).text
        counts = re.findall("\d+", count)
        number = int(counts[2]) // 10

        for page in range(1, number+2):
            address = current_url + '&pageNumber=' + str(page)
            driver.get(address)
            time.sleep(0.5)
            ps = driver.page_source
            soup = BeautifulSoup(ps,"html.parser")
            
            author = soup.select("div #cm_cr-review_list div.a-row.a-spacing-mini a.a-profile div.a-profile-content span.a-profile-name")
            star_rating = soup.findAll("i", {"data-hook": "review-star-rating"})
            date = soup.findAll("span", {"data-hook": "review-date"})
            title = soup.findAll("a", {"data-hook": "review-title"})
            comment = soup.findAll("span", {"data-hook": "review-body"})

            for a in author:
                comment_authors.append(a.text.strip('\n'))

            for s in star_rating:
                star_ratings.append(s.text.strip('\n')) 

            for d in date:
                comment_dates.append(d.text.strip('\n'))

            for t in title:
                comment_titles.append(t.text.strip('\n'))

            for c in comment:
                comments.append(c.text.strip('\n'))
        
        driver.close()

    except:
        pass

        driver.close()
```

```python
amazon_comment_result = pd.DataFrame( data = {'comment_author': comment_authors, 'star_rating': star_ratings, 'comment_date': comment_dates, 'comment_title' : comment_titles, 'comment' : comments}) ; amazon_comment_result  
```
